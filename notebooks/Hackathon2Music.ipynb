```python
# ==============================
# ðŸŽµ MUSIC ANALYTICS PIPELINE
# ==============================

# --- Step 1: Imports ---
import zipfile
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import spearmanr, kruskal

# --- Step 2: Configuration ---
zip_path = "../data/dataset.csv.zip"
extract_dir = "data_extracted"
eda_dir = "eda_outputs"
cleaned_path = "cleaned_music_data.csv"

for d in [extract_dir, eda_dir]:
    os.makedirs(d, exist_ok=True)

sns.set(style="whitegrid")
pd.set_option("display.max_columns", None)

# --- Step 3: Extract CSV ---
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

csv_files = [f for f in os.listdir(extract_dir) if f.endswith(".csv")]
if not csv_files:
    raise FileNotFoundError("No CSV found in extracted ZIP!")
csv_path = os.path.join(extract_dir, csv_files[0])

# --- Step 4: Load Dataset ---
try:
    df = pd.read_csv(csv_path, encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv(csv_path, encoding="latin1")

# --- Step 5: Detect key columns dynamically ---
track_col = next((c for c in df.columns if c.lower() in ["track_name", "track", "song", "title"]), df.columns[0])
artist_col = next((c for c in df.columns if c.lower() in ["artist_name", "artist", "artists"]), None)
genre_col = next((c for c in df.columns if c.lower() in ["genre", "genres"]), None)
pop_col = next((c for c in df.columns if c.lower() in ["popularity", "pop"]), None)
duration_col = next((c for c in df.columns if c.lower() in ["duration_ms", "duration"]), None)
release_col = next((c for c in df.columns if c.lower() in ["release_date", "year", "release"]), None)

# --- Step 6: Clean Data ---
df = df.drop_duplicates()
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
cat_cols = df.select_dtypes(exclude=np.number).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Duration in minutes
if duration_col:
    df["duration_min"] = df[duration_col] / 60000

# Release year
if release_col:
    df["release_year"] = pd.to_datetime(df[release_col], errors="coerce").dt.year

# --- Step 7: Standardize audio features ---
audio_features = [f for f in ['danceability','energy','loudness','acousticness','tempo','valence','instrumentalness'] if f in df.columns]
if audio_features:
    scaler = StandardScaler()
    df[audio_features] = scaler.fit_transform(df[audio_features])

# --- Step 8: Key Drivers of Popularity ---
if pop_col and audio_features:
    corr_pop = df[audio_features + [pop_col]].corr()[pop_col].sort_values(ascending=False)
    print("\nðŸ”¹ Correlation of Audio Features with Popularity:\n", corr_pop)

    plt.figure(figsize=(10,6))
    sns.heatmap(df[audio_features + [pop_col]].corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation plot")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/popularity_drivers_heatmap.png", dpi=300)
    plt.show()

# --- Step 9: Mood/Energy Classification ---
def classify_mood(row):
    if 'valence' in row and 'energy' in row:
        if row['valence'] > 0 and row['energy'] > 0:
            return "Happy/High Energy"
        elif row['valence'] <= 0 and row['energy'] > 0:
            return "Energetic/Sad"
        elif row['valence'] > 0 and row['energy'] <= 0:
            return "Chill/Happy"
        else:
            return "Sad/Low Energy"
    return "Unknown"

if all(f in df.columns for f in ['valence','energy']):
    df['mood_energy'] = df.apply(classify_mood, axis=1)

    plt.figure(figsize=(8,5))
    sns.countplot(y="mood_energy", data=df, palette="Set2", order=df['mood_energy'].value_counts().index)
    plt.title("Distribution of Songs by Mood & Energy")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/mood_energy_distribution.png", dpi=300)
    plt.show()

# --- Step 10: Genre-Level Analysis ---
if genre_col and audio_features:
    genre_summary = df.groupby(genre_col)[audio_features].mean()
    top_genres = genre_summary.sort_values('danceability', ascending=False).head(10)
    print("\nTop 10 genres by average danceability:\n", top_genres)

    plt.figure(figsize=(10,6))
    sns.heatmap(top_genres, annot=True, cmap="viridis")
    plt.title("Top Genres: Audio Feature Averages")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/genre_audio_features.png", dpi=300)
    plt.show()

# --- Step 11: Playlist Curation ---
if all(f in df.columns for f in ['energy','danceability']):
    playlist_high_energy = df[(df['energy'] > 0.5) & (df['danceability'] > 0.5)]
    display_cols = [track_col] + ([artist_col] if artist_col else []) + ['energy','danceability']
    display(playlist_high_energy[display_cols].head(10))

# --- Step 12: Boxplots for Popularity vs Nominal Variables (except track_genre) ---
nominal_cols = []
if genre_col:
    nominal_cols.append(genre_col)
if 'mood_energy' in df.columns:
    nominal_cols.append('mood_energy')

for col in nominal_cols:
    if col != "track_genre":  # Remove popularity vs track_genre
        plt.figure(figsize=(10,5))
        sns.boxplot(x=col, y=pop_col, data=df, palette="Set3")
        plt.title(f"Popularity vs {col}")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f"{eda_dir}/popularity_vs_{col}.png", dpi=300)
        plt.show()

# --- Step 13: Popularity per Artist (Average) ---
if artist_col and pop_col:
    # Top 20 by average popularity (all artists)
    artist_pop_avg = df.groupby(artist_col)[pop_col].mean().sort_values(ascending=False).head(20)
    plt.figure(figsize=(12,6))
    sns.barplot(x=artist_pop_avg.values, y=artist_pop_avg.index, palette="coolwarm")
    plt.title("Top 20 Artists by Average Track Popularity (All)")
    plt.xlabel("Average Popularity")
    plt.ylabel("Artist")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/avg_popularity_per_artist.png", dpi=300)
    plt.show()

    # Top 20 by average popularity (artists with â‰¥10 songs)
    artist_counts = df[artist_col].value_counts()
    eligible_artists = artist_counts[artist_counts >= 10].index
    artist_pop_avg_10plus = df[df[artist_col].isin(eligible_artists)].groupby(artist_col)[pop_col].mean().sort_values(ascending=False).head(20)
    plt.figure(figsize=(12,6))
    sns.barplot(x=artist_pop_avg_10plus.values, y=artist_pop_avg_10plus.index, palette="coolwarm")
    plt.title("Top 20 Artists by Average Popularity (â‰¥10 Songs)")
    plt.xlabel("Average Popularity")
    plt.ylabel("Artist")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/avg_popularity_per_artist_10plus.png", dpi=300)
    plt.show()

# --- Step 14: Binned Numeric Features vs Average Popularity ---
numeric_features = df.select_dtypes(include=np.number).columns.drop(pop_col)
for col in numeric_features:
    df[f'{col}_bin'] = pd.qcut(df[col], q=min(10, df[col].nunique()), duplicates='drop')
    avg_pop_by_bin = df.groupby(f'{col}_bin')[pop_col].mean().reset_index()
    
    plt.figure(figsize=(10,5))
    sns.barplot(x=f'{col}_bin', y=pop_col, data=avg_pop_by_bin, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.title(f"Average Popularity by {col} (Binned)")
    plt.xlabel(f"{col} bins")
    plt.ylabel("Average Popularity")
    plt.tight_layout()
    plt.savefig(f"{eda_dir}/{col}_binned_avg_popularity.png", dpi=300)
    plt.show()

# --- Step 15: Hypothesis Tests for Practical Significance ---
print("\nðŸ”¹ Step 15: Practical Significance Tests for All Features")

# Numeric features (Spearman correlation)
numeric_results = []
for col in numeric_features:
    corr, pval = spearmanr(df[col], df[pop_col], nan_policy='omit')
    numeric_results.append({
        'Feature': col,
        'Spearman_corr': corr,
        'p_value': pval,
        'Practical_Significant': abs(corr) >= 0.05
    })

numeric_df = pd.DataFrame(numeric_results).sort_values(by='Spearman_corr', key=abs, ascending=False)
print("\nNumeric Features:")
display(numeric_df)



# --- Step 16: Save Cleaned Data ---
df.to_csv(cleaned_path, index=False)
print(f"\nðŸ’¾ Cleaned dataset saved to: {cleaned_path}")
print(f"EDA visuals saved to: {eda_dir}/")








```
